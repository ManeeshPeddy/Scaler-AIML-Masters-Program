# How Fit, Transform and Fit_Transform works in sklearn?

### Fit

The `fit` method is about understanding, learning, or calculating properties from the training data that will be useful later, typically for transformation or prediction. It doesn't modify the data. For many scikit-learn objects, this is where the "learning" happens if it's a machine learning model, or where properties are computed if it's a transformer.


````python
from sklearn.impute import SimpleImputer
import numpy as np

data = [[1, 2], [np.nan, 3], [7, 6]]
imputer = SimpleImputer(strategy='mean')
imputer.fit(data)

# The imputer has now computed and stored the mean for each column.
print("Means:", imputer.statistics_)
````


### 2. `transform`
The `transform` method takes in data and returns a new version of the data after applying the transformation. The nature of the transformation depends on the specific algorithm or technique of the transformer or model.

#### SimpleImputer example with `transform`:
Using the previously `fit` imputer, you can now call `transform` on new data to fill missing values with the computed mean.


````python
new_data = [[np.nan, 10], [5, np.nan], [6, 7]]
filled_data = imputer.transform(new_data)

print(filled_data)
````



### 3. `fit_transform`
The `fit_transform` method essentially combines the `fit` and `transform` methods into one step. It's often more efficient to use `fit_transform` on the training data rather than using `fit` and `transform` separately. However, you should be cautious about using `fit_transform` on both training and testing data for certain preprocessing techniques, as it might cause data leakage.

#### SimpleImputer example with `fit_transform`:
Here, the imputer will calculate the mean values and then immediately use them to fill the missing values.



````python
data = [[1, 2], [np.nan, 3], [9, 100]]
imputer = SimpleImputer(strategy='mean')
filled_data = imputer.fit_transform(data)

print("Means:", imputer.statistics_)

print(filled_data)
````



### Summary:
- **`fit`**: Learns or computes properties from the data. Doesn't modify the data.
- **`transform`**: Applies a transformation or modification to the data based on what was learned during the `fit` step.
- **`fit_transform`**: Combines the two steps. Learns properties and immediately applies the transformation.

For transformers in scikit-learn, like `SimpleImputer`, the properties learned (e.g., column means) during the `fit` step are stored in instance variables (e.g., `imputer.statistics_`). This allows you to inspect what was learned, and it's also what allows the `transform` method to later apply transformations based on those learned properties.

# One Hot Encoding

One-hot encoding is a process of converting categorical data variables so they can be provided to machine learning algorithms to improve predictions. In one-hot encoding, each unique value of a categorical column is mapped to a new binary column, where each of these new columns represents a single unique value from the original column. In the new columns, the presence of the unique value is represented by a 1, while its absence is denoted by a 0. As a result, for a categorical column with n unique values, one-hot encoding will result in n new binary columns.

### Example 1: Simple DataFrame with a Single Categorical Column

**Original DataFrame:**

| Index | Color  |
|-------|--------|
| 0     | Red    |
| 1     | Green  |
| 2     | Blue   |
| 3     | Red    |

**One-hot encoded DataFrame:**

| Index | Color_Red | Color_Green | Color_Blue |
|-------|-----------|-------------|------------|
| 0     | 1         | 0           | 0          |
| 1     | 0         | 1           | 0          |
| 2     | 0         | 0           | 1          |
| 3     | 1         | 0           | 0          |

One-hot encoding can quickly increase the dimensionality of the dataset, especially if the categorical variables have many unique values. In cases where the categorical column has a large number of categories, the one-hot encoded data can become very sparse, meaning that the majority of the values in the resulting dataframe will be zeros. This sparsity can lead to increased memory consumption, slow computation, and sometimes can even degrade the performance of certain machine learning models.

### Example: High Cardinality Categorical Column

Suppose We've a dataset that records website visits, including a user ID and the domain they visited. Let's say you have only 5 users, but they've visited a variety of 1,000 unique websites.

**Original DataFrame (only a subset shown for simplicity):**

| Index | User_ID | Domain       |
|-------|---------|--------------|
| 0     | 1       | google.com   |
| 1     | 2       | yahoo.com    |
| 2     | 3       | wikipedia.org|
| ...   | ...     | ...          |
| 999   | 5       | example.com  |

If you decide to one-hot encode the 'Domain' column:

**One-hot encoded DataFrame (only a subset of columns shown for simplicity):**

| Index | User_ID | Domain_google.com | Domain_yahoo.com | Domain_wikipedia.org | ... | Domain_example.com |
|-------|---------|------------------|-----------------|----------------------|-----|-------------------|
| 0     | 1       | 1                | 0               | 0                    | ... | 0                 |
| 1     | 2       | 0                | 1               | 0                    | ... | 0                 |
| 2     | 3       | 0                | 0               | 1                    | ... | 0                 |
| ...   | ...     | ...              | ...             | ...                  | ... | ...               |
| 999   | 5       | 0                | 0               | 0                    | ... | 1                 |

This new DataFrame has over 1,000 columns (1 for each unique website) and most of them contain a value of '0' for each record. In the full encoded DataFrame of size (1,000 x 1,001), there would be 1,000,000 entries, but only 1,000 of those would be '1' — the rest would be '0'. This means that 99.9% of the data is just zeros. This kind of sparsity can be extremely inefficient in terms of memory usage and computational performance.

Sparse matrices, where the majority of entries are zeros, can be compressed to save space, but even with compression, extremely high dimensionality can challenge certain algorithms. For example, algorithms based on distance metrics (like k-NN) can perform poorly in high dimensions.

In such situations, alternative encoding strategies or dimensionality reduction techniques may be considered to handle high cardinality categorical columns efficiently.

# Label Encoding

Label encoding is another technique to transform categorical data into a format that machine learning algorithms can understand. Unlike one-hot encoding, which represents each category with a binary vector, label encoding represents each category with a unique number.

### Example 1: Simple DataFrame with a Single Categorical Column

**Original DataFrame:**

| Index | Color  |
|-------|--------|
| 0     | Red    |
| 1     | Green  |
| 2     | Blue   |
| 3     | Red    |

**DataFrame with Label Encoding:**

| Index | Color  | Color_encoded |
|-------|--------|---------------|
| 0     | Red    | 2             |
| 1     | Green  | 1             |
| 2     | Blue   | 0             |
| 3     | Red    | 2             |

**Mapping:**
- Red: 2
- Green: 1
- Blue: 0

### Example 2: DataFrame with Multiple Categorical Columns

**Mapping:**

Having both the original and encoded columns side by side will certainly make it easier for students to see the direct mappings.

It should be noted that the integer assignments for the categories might vary depending on the implementation and ordering of categories. The above examples illustrate one possible mapping.

It's worth noting that while label encoding is more space efficient than one-hot encoding, it can introduce ordinality to the data that might not actually exist. In the above examples, there's no inherent order to colors, but the model might infer that "Blue" (encoded as 0) is somehow "less than" "Green" (encoded as 1) and "Red" (encoded as 2). For certain algorithms, this can be problematic, so it's important to consider the nature of the data and the algorithm being used when deciding between label encoding and one-hot encoding.

# Target Encoding

Target encoding, also known as mean encoding, is another method to handle categorical variables. Instead of creating new columns like in one-hot encoding or giving arbitrary numbers like in label encoding, target encoding assigns each category in a categorical column with the average target value for that category.

### Implementation using `CategoryEncoders`:

`CategoryEncoders` is a Python library that provides several techniques for encoding categorical variables including target encoding.


````python
# label encode Loan_Status

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Loan_Status'] = le.fit_transform(data['Loan_Status'])

data.head()
````

````python
from category_encoders import TargetEncoder


pd.crosstab(data["Property_Area"], data["Loan_Status"], normalize="index")
````

target-encoder converts the categorical value in probability in terms of target-value.


It is also called as mean encoding. It is a supervised encoding technique, and based on the target value it assigns the probability of target value. It is also called as bin counting or likelihood encoding. It is similar to label encoding but in label encoding, we replace the categorical value with the number, whereas in target encoding we replace the categorical value with the probability of target value.

P[LS = 1 | PR = "RURAL"] OR P[pr = "RURAL" | ls = 1]



Remember, target encoding can introduce leakage if not done properly. Always encode the train and test sets separately or use techniques like cross-validation encoding to avoid potential pitfalls.


**Advantages**:
- Does not increase the dimensionality.
- Can incorporate the target information into encoding, potentially improving model performance.

**Disadvantages**:
- Prone to overfitting if not properly regularized or if the category has few occurrences.
- Introduces a relationship between the categorical column and the target, which may not always be desirable.

# Guidelines

---

### 1. **One-Hot Encoding**:

- **Description**: Converts each category value into a new column and assigns a 1 or 0 (True/False) value to the column.

- **When to Use**:
  - For nominal variables where no ordinal relationship exists.
  - With algorithms which expect orthogonal input features like linear regression.

- **Example**: For a column `Color` with values `Red`, `Blue`, and `Green`, three new columns will be created: `is_Red`, `is_Blue`, and `is_Green`. A record with `Color = Red` would have `is_Red = 1` and `is_Blue = 0` and `is_Green = 0`.

- **Advantages**:
  - Easy to interpret.
  - No introduction of arbitrary numbers, which prevents potential misinterpretation by the model.

- **Disadvantages**:
  - Can increase data dimensionality rapidly if the category has many unique values.

---

### 2. **Label Encoding**:

- **Description**: Converts each category value into a unique integer.

- **When to Use**:
  - For ordinal variables where a hierarchy or order is inherent.
  - With tree-based algorithms (like decision trees and random forests) which can interpret the ordinal nature.

- **Example**: For a column `Size` with values `Small`, `Medium`, and `Large`, these could be encoded as `0`, `1`, and `2` respectively.

- **Advantages**:
  - Efficient in terms of memory as only one column is needed.
  - Can capture the ordinal nature of certain variables.

- **Disadvantages**:
  - Introduces an arbitrary order for nominal variables.

---

### 3. **Target (Mean) Encoding**:

- **Description**: Uses the mean of the target variable for each category as its new value.

- **When to Use**:
  - When a categorical variable has high cardinality.
  - When there's a correlation between the category and the target variable.
  - Use cautiously with tree-based models.

- **Example**: If you're predicting house prices and have a categorical column `Neighborhood`, houses in the `Beverly Hills` neighborhood might have an average price of `$1M`, which becomes the encoded value for that category.

- **Advantages**:
  - Efficiently captures the relationship between category and target variable, especially for high cardinality categories.
  - Reduces dimensionality.

- **Disadvantages**:
  - Can lead to overfitting.
  - Potential for data leakage if not done properly.
  - Might introduce collinearity.

---

**Key Recommendations**:

1. **Data Splitting**: Always split your dataset into training and test sets before any encoding. This prevents information from the test set leaking into your training set.

2. **Algorithm Type**: Consider the nature of your algorithm. Linear and distance-based algorithms interpret encoded data differently compared to tree-based algorithms.

3. **Experiment**: If in doubt, experiment with different encodings and compare model performance.

4. **Be Cautious**: Particularly with target encoding, ensure you're not introducing data leakage or overfitting.

5. **Deep Learning**: For those using neural networks, consider embeddings for categorical data. It's a powerful technique, particularly for categories with many unique values.
